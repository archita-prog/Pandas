```python
import pandas as pd
df = pd.read_csv("weather_data.csv", parse_dates = ["day"])
df
```

```python
df.set_index('day', inplace = True)
df
```

```python
new_df = df.fillna(0)
new_df
```

```python
new_df = df.fillna({
    'temperature': 0,
    'windspeed': 0,
    'event': 'no event'
})
new_df
```

```python
new_df = df.interpolate(methid = "time")
new_df
```

```python
new_df = df.dropna(how = "all")
new_df
```

```python
new_df = df.dropna(thresh = 1)
new_df
```
