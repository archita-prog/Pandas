```python
import pandas as pd
df1 = pd.DataFrame({
    "city": ["new york", "chicago","orlando"],
    "temperature" : [21, 14,25],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["new york", "chicago","maryland"],
    "humidity" : [65, 68,75],
})
df2
```

```python
df3 = pd.merge(df1, df2, on = "city", how = "outer")
df3
```

```python
df3 = pd.merge(df1, df2, on = "city")
df3
```

```python
df3 = pd.merge(df1, df2, on= "city", how = "outer", indicator = True)
df3
```
